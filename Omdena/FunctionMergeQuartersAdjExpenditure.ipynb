```python
def merge_newids(df, column_list):    
    
    result = pd.DataFrame(columns = column_list, index = ['myindex']) ## TODO: keep index information
    
    
    ######################### generating a list of unique NEWIDs ##############################
    unique_NEWID = df[df.duplicated(['NEWID'])]

    for elem in unique_NEWID.loc[:, 'NEWID']:
        
        dfff = df[df.NEWID == elem]
        
        temp_df = dfff.loc[:, 'NEWID':'Year']
        duplicateRowsDFDDD = temp_df[temp_df.duplicated()]
        quarter = 0
        
        ######
        for i in range (dfff.shape[0]):
    
            if dfff.iat[i,22] != 'x':
        
                quarter = dfff.iat[i,22]

        duplicateRowsDFDDD['Quarter'] = quarter
        
        ##### summing up the values from this and last quarter
        
        for val in duplicateRowsDFDDD.columns[1:21]:

            result.loc[:, val] = duplicateRowsDFDDD[val].sum()
            #duplicateRowsDFDDD[elem].sum()
            
        result.loc[:, 'NEWID'] = duplicateRowsDFDDD.iat[0,0]
        result.loc[:, 'Year'] = duplicateRowsDFDDD.iat[0,21]
        result.loc[:, 'Quarter'] = duplicateRowsDFDDD.iat[0,22]
    
    return(result)
```
